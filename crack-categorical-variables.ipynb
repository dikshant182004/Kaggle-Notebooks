```python
# so LET ME INTRODUCE YOU ALL WITH THE RULES OF THIS WORLD 
# a simple dataset in which there are various categorical variables which contains high cardinality in imagination 
# This dataset is mere small but it will give you the rough idea about high cardinality data.
import pandas as pd
import numpy as np

np.random.seed(42)

category_1 = np.random.choice(['Category_' + str(i) for i in range(1, 21)], 60)

category_2 = np.random.choice(['Type_' + str(i) for i in range(1, 16)], 60)

category_3 = np.random.choice(['Group_' + str(i) for i in range(1, 11)], 60)

category_4 = np.random.choice(['Item_' + str(i) for i in range(1, 26)], 60)

target = np.random.randint(100, 1000, 60)

category_5 = np.random.choice(['Label_' + str(i) for i in range(1, 31)], 60)

df = pd.DataFrame({
    'Category_1': category_1,
    'Category_2': category_2,
    'Category_3': category_3,
    'Category_4': category_4,
    'Category_5': category_5,
    'Target': target
})
```

```python
# split the data to determine the result of encoding techniques on model accuracy
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['Target']),df['Target'],
                                              test_size=0.2,random_state=5)
```

 # **Target encoding ** (also known as mean encoding)
- **IN SHORT** ,it Replace each category with the mean of the target variable for that category.
- **For example** - > there is a City column with purchase as target column so lets imagine there is a city A which appears 5 times in dataset with target(purchase) column values as (1,2,3,4,5) {sounds cheap but they are in crores} so the city categorical value will get replace with the mean of its target values which is 3 (simple and sweet).


```python
from category_encoders.target_encoder import TargetEncoder
ans_1 = TargetEncoder(cols = X_train.columns).fit(X_train, y_train)
X_train_encoded = ans_1.transform(X_train)
X_test_encoded = ans_1.transform(X_test)

# To apply smoothing for preventing overfitting
# ans_1 = TargetEncoder(cols = X_train.columns,smoothing=10).fit(X_train, y_train) 
```

```python
X_train_encoded.sample(2)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train_encoded, y_train)

y_pred = model.predict(X_test_encoded)
```

```python
from sklearn.metrics import mean_squared_error

# Just calculating MSE for simplicity
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

PROS:
- parameter free
- Reduced Dimensionality: Unlike one-hot encoding, which can create many features for high-cardinality categorical variables, target encoding reduces the number of features.
- Preserves Information: It captures the relationship between the categorical feature and the target variable, potentially improving model performance.
CONS:
- risk of target leakage (target leakage means using some information from target to predict the target itself)
- Overfitting Risk: Since target encoding uses the target variable to encode the categorical feature, there is a risk of overfitting, especially when the dataset is small or when a category has very few observations.
USES:
- Works well with categorical variables that have a strong relationship with the target variable.

# CROSS VALIDATION 
- Just Spliting  the data into k folds. For each fold, calculating  the mean target for the training set and encode the validation set.

****CROSS VALIDATION****
- Just Spliting  the data into k folds. For each fold, calculating  the mean target for the training set and encode the validation set.
-
****SMOOTHING****
- Combine the category mean with the overall mean of the target variable by using the formula 
- weighted by the number of observations in each category and it determined by smoothing factor as discussed above.
-
****RANDOM NOISE****
- Adding a small amount of random noise(errors) to the encoded values can help to prevent the model from overfitting to the exact target values.
